```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books_df = pd.read_csv("Data_books/all_books.csv")
books_df.head(15)
```

```python
# Count the number of missing value each column
missing_percent = books_df.isnull().mean()
# Plot the missing value percentages
missing_percent.plot.bar(stacked=True, figsize=(8,4))
plt.title("The percentage of missing value in each column")
plt.xlabel("Column Name")
plt.ylabel("Percentage of missing value")
plt.show()
```

```python
plt.figure(figsize=(10,4))
ax = sns.countplot(books_df, x= "year_first_published")
ax.bar_label(ax.containers[0])
ax.axes.set_title("The number of books published in year",fontsize=9)
ax.set_xlabel("First year published",fontsize=9)
ax.set_ylabel("Number of books", fontsize=9)
plt.show()
```

```python
gd = sns.lmplot(books_df, x = 'average_rating', y = 'num_pages', line_kws={'color':'r'},height=4, aspect=2)
plt.title("The correlation between average rating and number of book's pages", fontsize=9)
plt.xlabel("Average Rating", fontsize=9)
plt.ylabel("Number of pages", fontsize=9)
plt.show()
```
