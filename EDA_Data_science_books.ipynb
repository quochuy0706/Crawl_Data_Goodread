```python
import re

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books_df = pd.read_csv("Data_books/all_books.csv")
books_df.head(15)
```

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(books_df, x= "year_first_published")
ax.bar_label(ax.containers[0])
ax.axes.set_title("The number of books published in year",fontsize=9)
ax.set_xlabel("First year published",fontsize=9)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax = sns.scatterplot(books_df, x='average_rating', y = 'num_pages')
ax.axes.set_title("The correlation between average rating and number of book's pages",fontsize=9)
ax.set_xlabel("Average Rating",fontsize=9)
ax.set_ylabel("Number of pages",fontsize=9)
plt.show()
```
